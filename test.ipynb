```python
import pandas as pd

# Data for drinks
data = {
    "Drink": [
        "Espresso", "Americano", "Cappuccino", "Latte", "Drip/Filter Coffee", "Instant Coffee",
        "Thai Iced Coffee (Gafae Yen)", "Black Tea", "Green Tea", "Oolong Tea", "White Tea", "Matcha",
        "Thai Iced Tea (Cha Yen)", "Red Bull", "Monster", "M-150", "Carabao", "5-hour Energy"
    ],
    "Serving Size (ml)": [
        30, 240, 240, 240, 240, 240, 240, 240, 240, 240, 240, 240, 240, 250, 500, 250, 330, 57
    ],
    "Caffeine per Serving (mg)": [
        63, 95, 75, 75, 130, 100, 70, 55, 32, 40, 22, 70, 45, 80, 160, 80, 106, 200
    ],
    "Caffeine per Litre (mg)": [
        2100, 396, 313, 313, 542, 417, 292, 229, 133, 167, 92, 292, 188, 320, 320, 320, 321, 3509
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set style
sns.set_style("whitegrid")

# Horizontal bar plot
plt.figure(figsize=(12,8))
sns.barplot(data=df, x="Caffeine per Litre (mg)", y="Drink", palette="viridis")

# Add title and labels
plt.title("Caffeine Content per Litre in Different Drinks", fontsize=16)
plt.xlabel("Caffeine per Litre (mg)", fontsize=14)
plt.ylabel("Drink", fontsize=14)

# Show exact values on bars
for index, value in enumerate(df["Caffeine per Litre (mg)"]):
    plt.text(value + 20, index, str(value), va='center')

plt.tight_layout()
plt.show()
```
